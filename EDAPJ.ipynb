```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


titanic=sns.load_dataset('titanic')

print(titanic.info())
df=pd.DataFrame(titanic)
print(df.shape)
print(df.isnull().sum())

df['age']=df['age'].fillna(df['age'].mean())
df['fare']=df['fare'].fillna(df['fare'].median())
df['embarked']=df['embarked'].fillna("Unknown")
print(df.dtypes)
df['age']=df['age'].astype(int)
df['FamilySize']=df['sibsp']+df['parch']+1

plt.figure(figsize=(14,32))

plt.subplot(5,1,1)
sex_survival=df.groupby('sex',observed=False)['survived'].mean()
plt.bar(sex_survival.index,sex_survival.values,color='skyblue')
plt.ylabel("Survival Rate")
plt.xlabel("Sex")
plt.title("Gender Vs Survival")

plt.subplot(5,1,2)
class_survival=df.groupby('class',observed=False)['survived'].mean()
plt.bar(class_survival.index,class_survival.values,color='skyblue')
plt.ylabel("Survival Rate")
plt.xlabel("class")
plt.title("Class Vs Survival")

plt.subplot(5,1,3)
sns.histplot(df['age'],bins=20,kde=True,color='orange')
plt.title("Age Distribution of titanic")
plt.xlabel("Age")
plt.ylabel("Count")

plt.subplot(5,1,4)
sns.boxplot(x='fare',y='class',data=df)
plt.title("Fare-Class distribution of titanic")

plt.subplot(5,1,5)
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm",linewidths=0.5)
plt.title("Titanic Correlation Map")

sns.pairplot(df[['age','fare','survived']],hue='survived')
plt.suptitle("Pairplot of titanic Features",y=1.02)

print("Key Insights:")
print("1. Females had higher survival rates than males.")
print("2. First-class passengers had the best survival chances.")
print("3. Younger passengers survived more often.")
print("4. Higher fare correlates with survival.")


```
